```python
import pandas as pd

# Read the CSV file
df = pd.read_csv("Final.csv")

# Display the dataframe
print(df)
```

```python
label_counts = df['label'].value_counts()

print("Different labels and their total numbers:")
for label, count in label_counts.items():
    print(f"Label: {label}, Count: {count}")
```

```python
# Check for null values in the DataFrame
null_counts = df.isnull().sum()
print("Null Value Counts:")
print(null_counts)
```

```python
from sklearn.model_selection import train_test_split

# Separate features and target variable
X = df['title']
y = df['label']

# Stratify the data split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Combine the features and target variable for training and validation sets
train_df = pd.concat([X_train, y_train], axis=1)
val_df = pd.concat([X_val, y_val], axis=1)

print('trainset size:', train_df.shape)
print('valset size:', val_df.shape)
print('trainset label distribution:', train_df['label'].value_counts())
print('valset label distribution:', val_df['label'].value_counts())
```

```python
# Save the training DataFrame as train.tsv
train_df.to_csv('/content/drive/My Drive/dataset/train.tsv', sep='\t', index=False)
# Save the validation DataFrame as val.tsv
val_df.to_csv('/content/drive/My Drive/dataset/val.tsv', sep='\t', index=False)
```
